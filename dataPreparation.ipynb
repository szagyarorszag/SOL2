```python
import arxiv
import pandas as pd
import time
from tqdm import tqdm  
arxiv_categories = [
    "physics", "math", "cs", "q-bio", "q-fin", "stat",
    "astro-ph", "cond-mat", "gr-qc", "hep-ex", "hep-lat",
    "hep-ph", "hep-th", "math-ph", "nlin", "nucl-ex",
    "nucl-th", "physics-app", "physics-atm", "physics-bio",
    "physics-cl", "physics-chem", "physics-dae", "physics-ed",
    "physics-foc", "physics-gen", "physics-hall", "physics-med",
    "physics-optics", "physics-plasm", "physics-pop-ph",
    "physics-soc-ph", "physics-space-ph"
]

def create_search_query(category):
    return f"cat:{category}"

# Parameters
max_results_per_category = 1000
batch_size = 100
sleep_time = 3 
all_data = []

for category in tqdm(arxiv_categories, desc="Categories"):
    try:
        search = arxiv.Search(
            query=create_search_query(category),
            max_results=max_results_per_category,
            sort_by=arxiv.SortCriterion.SubmittedDate
        )
        
        for result in search.results():
            all_data.append({
                'id': result.get_short_id(),
                'title': result.title,
                'authors': ', '.join([author.name for author in result.authors]),
                'abstract': result.summary,
                'categories': ', '.join(result.categories),
                'published': result.published.strftime('%Y-%m-%d'),
                'updated': result.updated.strftime('%Y-%m-%d'),
                'pdf_url': result.pdf_url,
                'arxiv_url': result.entry_id
            })
        
        print(f"Fetched {len(all_data)} records so far...")

        time.sleep(sleep_time)
    
    except Exception as e:
        print(f"An error occurred while processing category '{category}': {e}")
        continue

df = pd.DataFrame(all_data)

print(df.head())

df.to_csv('arxiv_large_diverse_dataset.csv', index=False)
print("Data saved to 'arxiv_large_diverse_dataset.csv'")
```

```python
import pandas as pd

column_names = [
    'arxiv_id',
    'title',
    'authors',
    'abstract',
    'categories',
    'submission_date',
    'last_revision_date',
    'pdf_url',
    'abs_url'
]

try:
    df = pd.read_csv(
        'arxiv_large_diverse_dataset.csv',
        names=column_names,
        header=0,
        quotechar='"',
        escapechar='\\',
        engine='python',
        on_bad_lines='skip'
    )
    print("Dataset loaded successfully, with some lines skipped due to formatting issues.")
except Exception as e:
    print(f"Error loading dataset: {e}")
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

df.fillna('', inplace=True)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = nltk.word_tokenize(text.lower())
    
    tokens = [word for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(tokens)

df['processed_abstract'] = df['abstract'].apply(preprocess_text)
print("Text preprocessing completed.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer(max_features=5000)  # Adjust max_features as needed

tfidf_matrix = tfidf_vectorizer.fit_transform(df['processed_abstract'])
print("TF-IDF vectorization completed.")

cosine_sim_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
print("Cosine similarity matrix computed.")
```

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=300, random_state=42)
reduced_tfidf = svd.fit_transform(tfidf_matrix)
print("Dimensionality reduction completed.")
```

```python
import networkx as nx

G = nx.Graph()

for index, row in df.iterrows():
    G.add_node(row['arxiv_id'], title=row['title'], authors=row['authors'], abstract=row['abstract'])

print(f"Added {G.number_of_nodes()} nodes to the graph.")

similarity_threshold = 0.2

for i in range(len(df)):
    for j in range(i+1, len(df)):
        similarity = cosine_sim_matrix[i][j]
        if similarity >= similarity_threshold:
            G.add_edge(df.at[i, 'arxiv_id'], df.at[j, 'arxiv_id'], weight=similarity)

print(f"Added {G.number_of_edges()} edges to the graph.")
```

```python
for index, row in df.iterrows():
    citing_paper = row['arxiv_id']
    cited_papers = row['citations'].split(';') if 'citations' in df.columns else []
    for cited_paper in cited_papers:
        if cited_paper in df['arxiv_id'].values:
            G.add_edge(citing_paper, cited_paper, weight=1.0)

print(f"Updated graph with citation-based edges. Total edges: {G.number_of_edges()}")
```

```python
import networkx as nx
from networkx.algorithms import community
import matplotlib.pyplot as plt

G = nx.karate_club_graph()

girvan_newman = community.girvan_newman(G)

top_level_communities = next(girvan_newman)
communities = sorted(map(sorted, top_level_communities))

partition = {}
for community_id, community_nodes in enumerate(communities):
    for node in community_nodes:
        partition[node] = community_id

nx.set_node_attributes(G, partition, 'community')

print(f"Detected {len(communities)} communities in the graph.")

pos = nx.spring_layout(G)
colors = [partition[node] for node in G.nodes()]

plt.figure(figsize=(10, 7))
nx.draw_networkx_nodes(G, pos, node_size=300, node_color=colors, cmap=plt.cm.Set3)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12)
plt.title("Girvan-Newman Community Detection")
plt.axis('off')
plt.show()
```

```python
import networkx as nx
from networkx.algorithms import community
import matplotlib.pyplot as plt

G = nx.karate_club_graph()

girvan_newman = community.girvan_newman(G)
girvan_first_level = next(girvan_newman)
girvan_communities = sorted(map(sorted, girvan_first_level))

label_prop_communities = sorted(map(sorted, community.label_propagation_communities(G)))

greedy_communities = sorted(map(sorted, community.greedy_modularity_communities(G)))

def assign_partition(communities):
    partition = {}
    for community_id, community_nodes in enumerate(communities):
        for node in community_nodes:
            partition[node] = community_id
    return partition

partition_girvan = assign_partition(girvan_communities)
partition_label = assign_partition(label_prop_communities)
partition_greedy = assign_partition(greedy_communities)

pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
colors = [partition_girvan[node] for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=300, node_color=colors, cmap=plt.cm.Set1)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12)
plt.title("Girvan-Newman")

plt.subplot(1, 3, 2)
colors = [partition_label[node] for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=300, node_color=colors, cmap=plt.cm.Set2)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12)
plt.title("Label Propagation")

plt.subplot(1, 3, 3)
colors = [partition_greedy[node] for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_size=300, node_color=colors, cmap=plt.cm.Set3)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12)
plt.title("Greedy Modularity Maximization")

plt.tight_layout()
plt.show()

print(f"Girvan-Newman detected {len(girvan_communities)} communities.")
print(f"Label Propagation detected {len(label_prop_communities)} communities.")
print(f"Greedy Modularity Maximization detected {len(greedy_communities)} communities.")
```
